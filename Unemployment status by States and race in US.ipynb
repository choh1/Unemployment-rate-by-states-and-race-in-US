```python
#rename column name
df_asian.rename(columns={"LNU04032183":"asian"}, inplace=True) 
df_hsla.rename(columns={"LNS14000009":"hispanic"},inplace=True) 
df_white.rename(columns={"LNS14000003":"white"},inplace=True) 
df_black.rename(columns={"LNS14000006":"black"},inplace=True) 
```

```python
#changing date format from object to datetime
df_asian['DATE'] = pd.to_datetime(df_asian['DATE'], format='%Y-%m-%d')
df_hsla['DATE'] = pd.to_datetime(df_hsla['DATE'], format='%Y-%m-%d')
df_white['DATE'] = pd.to_datetime(df_white['DATE'], format='%Y-%m-%d')
df_black['DATE'] = pd.to_datetime(df_black['DATE'], format='%Y-%m-%d')
```

```python
#plot
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=(10,10)

plt.plot(df_asian['DATE'], df_asian['asian'], color='black', label='asian')
plt.plot(df_hsla['DATE'],df_hsla['hispanic'], color='yellow', label='hispanic')
plt.plot(df_white['DATE'],df_white['white'], color='blue', label='white')
plt.plot(df_black['DATE'],df_black['black'], color='pink', label='black')

plt.legend(loc='upper left')
plt.xlabel='Date'
plt.ylabel='Unemployment Rate'
plt.show
```
